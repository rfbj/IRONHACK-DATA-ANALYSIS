# In this lab you will use the selenium library to scrape and interact with dynamic websites

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time 
from selenium.webdriver.chrome.service import Service
chrome_path = "C:\\chromedriver.exe" #mudar estrutura
ser = Service(chrome_path)
driver = webdriver.Chrome(service=ser)
```

# Hackathon 
## Find the next event happening in your area 

```python
from PIL import Image 
url_hack = 'https://www.eventbrite.com.br/d/online/hackathon/'
driver.get(url_hack)
inp_box = driver.find_element("css selector","#locationPicker")
inp_box.send_keys("SÃ£o Paulo")
time.sleep(1)
find_but = driver.find_element("css selector","#locationPicker-listbox > li > div > button > div > div > div")
time.sleep(1)
find_but.click()
time.sleep(1)
find_but2 = driver.find_element("css selector","#root > div > div.eds-structure__body > div > div > div > div.eds-fixed-bottom-bar-layout__content > div > main > div > div > aside > div > div > ul > li:nth-child(4) > button")
time.sleep(1)
find_but2.click()
time.sleep(1)
find_hj = driver.find_element("css selector", "#root > div > div.eds-structure__body > div > div > div > div.eds-fixed-bottom-bar-layout__content > div > main > div > div > aside > div > div > div > div:nth-child(3) > button")
time.sleep(1)
find_hj.click()
time.sleep(1)
course_data = driver.find_element("css selector","#root > div > div.eds-structure__body > div > div > div > div.eds-fixed-bottom-bar-layout__content > div > main > div > div > section.search-base-screen__search-panel > div.search-results-panel-content > div")
time.sleep(1)
nxt_course = pd.DataFrame(course_data.text.split("\n")).T
nxt_course
nxt_course.columns=["Name","Time","Location","Investment","Site","Followers","Status"]
```
