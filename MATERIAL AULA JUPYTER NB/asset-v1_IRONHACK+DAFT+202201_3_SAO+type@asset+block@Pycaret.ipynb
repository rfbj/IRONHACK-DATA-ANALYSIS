## Import libraries, pandas and pycaret.classification

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from pycaret.classification import *
```

## Read and separate data

```python
df = pd.read_csv('data/loan_dataset_train.csv')
```

```python
data_train,data_test = train_test_split(df, test_size = 0.1)
```

## Pycaret magic! setup the pipeline

```python
classification = setup(data = data_train, target = 'Loan_Status', session_id=42) 
```

## Compare classification models

```python
best_model = compare_models()
```

```python
print(best_model)
```

```python
et = create_model('et')
```

```python
plot_model(et, plot = 'confusion_matrix')
```

```python
plot_model(et, plot = 'auc')
```

```python
plot_model(et, plot='feature')
```

```python
final_et = finalize_model(et)
```

```python
predict_model(final_et, data=data_test)
```
